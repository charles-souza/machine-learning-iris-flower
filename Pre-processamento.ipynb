<h4>Carregando Bibliotecas<h4/>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

<h4>Carregando o CSV em um DataFrame<h4/>

```python
df = pd.read_csv('iris.csv')
```

<h4>Selecionado as "features" para classificação em um array NUMPY<h4/>

```python
X = np.array(df.drop('target', 1))
X
```

<h4>Selecionado as classes para classificação em um array NUMPY<h4/>

```python
Y = np.array(df.target)
Y
```

Classificador KNN

<h4>Importando KNN do scikit-learn<h4/>

```python
from sklearn.neighbors import KNeighborsClassifier
```

<h4>Criando um classificador<h4/>

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

<h4>Treinando o classificador<h4/>

```python
knn.fit(X,Y)
```
